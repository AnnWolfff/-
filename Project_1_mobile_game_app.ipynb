**Функция для подсчета retention игроков по дням от даты регистрации игрока:**

```python
def retention(period_start, data=df, user_id='uid'):
    """Принимает дату начала периода, возвращает retention по дням в виде графика (heatmap)"""
    
    data = data.query('reg_date >= @period_start')
    
    data['reg_date'] = pd.to_datetime(data['reg_date']).dt.to_period("d")
    data['auth_date'] = pd.to_datetime(data['auth_date']).dt.to_period("d")
    
    #создаем колонку разницы в днях между ключевым действием (заход пользователя в игру) и регистрацией 
    data['days'] = (data['auth_date'] - data['reg_date']).apply(attrgetter('n'))
    
    # группируем и считаем число уникальных пользователей
    data_df = data.groupby(['reg_date', 'days'], as_index=False).agg({user_id: 'nunique'})
    
    retention_df = data_df.pivot(index='reg_date', columns='days', values=user_id)
    
    # делим все значения на число пользователей из первой колонки
    retention_df = retention_df.div(retention_df[0], axis=0)
    
    #устанавливаем параметры графика
    sns.set_theme(style='white', font_scale=2)
    plt.figure(figsize=(50, 20))
    plt.title('Когортный анализ retention')
    
    sns.heatmap(retention_df, mask=retention_df.isnull(), annot=True, fmt='.2%', vmin=0.0, vmax=0.1, cmap='RdYlGn')
    
    print(retention.__doc__)
```

```python
retention(period_start="2020-09-09")
```

```python
df_userCount = df.groupby('testgroup', as_index=False) \
                .agg({'user_id': 'count'}) \
                .rename(columns={'user_id': 'full_count_users'})
df_userCount
```

```python
df_userCount_revenue_not0 = df.query('revenue != 0') \
                            .groupby('testgroup', as_index=False) \
                            .agg({'user_id': 'count'}) \
                            .rename(columns={'user_id': 'count_buyers'})
df_userCount_revenue_not0
```

```python
DF = df_userCount.merge(df_userCount_revenue_not0, how="inner", on='testgroup')

DF['count_notbuyers'] = DF.full_count_users - DF.count_buyers
DF['CR'] = round((DF.count_buyers / DF.full_count_users) * 100, 2)

DF
```

```python
delta = (DF.CR.loc[0] - DF.CR.loc[1]).round(2)
delta 
```

_Конверсия в группе **а** на 0.06 больше, чем в тестовой группе **b**_

_Для определения стат значимости данных различий применим критерий Хи-квадрат_ \
_H0: CR не зависит от набора акционных предложений_

```python
DF_chi2 = DF[['count_buyers', 'count_notbuyers', 'testgroup']].set_index('testgroup')
DF_chi2
```

```python
chi2_contingency(DF_chi2)
```

```python
chi2_contingency(DF_chi2)[1] < 0.05
```

_pvalue < 0.05 => отвергаем H0; различия стат значимы_
